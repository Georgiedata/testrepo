## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
df = pd.read_sql_query("SELECT * FROM master", conn)
df
plt.hist(df["ConvertedComp"])
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
import seaborn as sns
sns.boxplot(df["Age"])
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
df.plot(kind='scatter', x= 'Age', y='WorkWeekHrs', figsize=(10, 4), color='red')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
```

## Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import plotly.express as px

px.scatter(df, x='WorkWeekHrs', y='CodeRevHrs', color='Age', size_max=60)
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

query="""SELECT DatabaseDesireNextYear, Count(*) as count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY DatabaseDesireNextYear DESC LIMIT 5
"""

DatabaseDesireNextYear = pd.read_sql_query(query, conn)
x = DatabaseDesireNextYear['count']
label = DatabaseDesireNextYear['DatabaseDesireNextYear']
plt.pie(x, labels = label, autopct = '%1.1f%%', radius = 1.6)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

query="""SELECT WorkWeekHrs, CodeRevHrs, Age FROM
MASTER WHERE Age BETWEEN 30 and 35
"""

df1= pd.read_sql_query(query, conn)
x = df1.groupby("Age").median()
x.plot(kind="bar", stacked=True)
plt.title('WorkWeekHrs and CodeRevHrs for age 30-35')
plt.xlabel("Age")
plt.ylabel("Hours")
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY="""SELECT ConvertedComp, Age FROM
MASTER WHERE Age BETWEEN 45 AND 60"""
y=pd.read_sql_query(QUERY, conn)
y1=y.groupby(y.Age)["ConvertedComp"].median()
plt.subplots(figsize=(20,6))
y1.plot(kind='line')
plt.xlabel("ConvertedComp")
plt.ylabel("Age")

```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY="""SELECT MainBranch, count(MainBranch) FROM
MASTER
group by MainBranch
"""
y=pd.read_sql_query(QUERY,conn)
y.plot(kind="barh", figsize=(12,5))
plt.xlabel("MainBranch")
plt.show()
```
